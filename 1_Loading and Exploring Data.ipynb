```python
df_cat= df[['workclass', 'education', 'marital-status', 'occupation' , 'relationship', 'race', 'sex', 'native-country', 'salary']]
    
df_cat['education'].value_counts().plot.bar()
```

```python
#Alternative historgramm with sns-library
sns.distplot(df['hours-per-week'])
```

```python
sns.distplot(df['age'])
```

```python
#box plot relationship with categorical attributes
var = 'education'
data = pd.concat([df['capital-gain'], df[var]], axis=1)
f, ax = plt.subplots(figsize=(12, 6))
fig = sns.boxplot(x=var, y="capital-gain", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
# good for non-categorical data
sns.scatterplot( x = "hours-per-week", y = "capital-gain", data = df, hue="salary")
```

```python
#Overview of various plot combinations
sns.pairplot(df, hue="salary", height=3)
```

```python
# Display a correlation matrix to discover possible relationships in the numeric variables

df_num = df.select_dtypes(include=['int64']) # select all variables with integers 

plt.figure(figsize=(16,6))
mask = np.triu(df_num.corr(),k=1)
sns.heatmap(df_num.corr(), annot=True, linewidths=1, mask=mask, cmap='coolwarm')

plt.title('Correlation Matrix')
plt.show()
```
